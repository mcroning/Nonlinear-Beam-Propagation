```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
from scipy import ndimage
data=np.load('/content/drive/MyDrive/2bcdynamicim/data.npz',allow_pickle=True)
prdata=data["arr_0"].item()
ixzf=data["arr_2"]
#imin_final=ndimage.rotate(data["arr_3"],270)
#imout_final=ndimage.rotate(data["arr_4"],270)
imin_final=data["arr_3"]
imout_final=data["arr_4"]
fanning_study=True
try:
  fanout=data["arr_5"]
except:
  fanning_study=False
```

```python
xaper=prdata['xaper']
yaper=prdata['yaper']
xsamp=prdata['xsamp']
ysamp=prdata['ysamp']
gl=prdata['gl']
```

```python
ymax=yaper/2/2 # in mm 1st /2 is for half ya aperture spanning, 2nd /2 is for zero centered yaper
fig,ax=plt.subplots()
ax.set_title(r'input image with gain $\gamma$l '+str(gl))
ax.set_xlabel(r'x ($\mu$m)')
ax.set_ylabel(r'y ($\mu$m)')
img=ax.imshow(imin_final.T,extent=[-ymax,ymax,-ymax,ymax])
plt.colorbar(img)
plt.show()
```

```python
ymax=yaper/2/2 # in mm 1st /2 is for half ya aperture spanning, 2nd /2 is for zero centered yaper
fig,ax=plt.subplots()
ax.set_title(r'output image with gain $\gamma$l '+str(gl))
ax.set_xlabel(r'x ($\mu$m)')
ax.set_ylabel(r'y ($\mu$m)')
img=ax.imshow(imout_final.T,extent=[-ymax,ymax,-ymax,ymax],vmax=1)
plt.colorbar(img)
plt.show()
```

```python
fig,ax=plt.subplots()
ixzf=np.where(ixzf<0.0001,0.0001,ixzf)
ax.set_title(r'xz cross section $\gamma$l '+ str(gl))
ax.set_xlabel(r'x ($\mu$m)')
ax.set_ylabel(r'z ($\mu$m)')
img=ax.imshow(ixzf,norm=colors.LogNorm(vmin=0.01,vmax=ixzf.max()),extent=[-xaper/2,xaper/2,prdata['rlen'],0],aspect=1)
plt.colorbar(img)
plt.show()
```

```python
if fanning_study==True:
  fymax=ysamp/yaper/2
  fxmax=xsamp/xaper/2
  fanout=fanout/np.max(fanout)
  fig,ax=plt.subplots()
  ixzf=np.where(ixzf<0.0001,0.0001,ixzf)
  ax.set_title(r'far field $\gamma$l '+ str(gl))
  ax.set_xlabel(r'x ($\mu$m)')
  ax.set_ylabel(r'z ($\mu$m)')
  img=ax.imshow(fanout.T,extent=[-fxmax,fxmax,-fymax,fymax],cmap='terrain')
  plt.colorbar(img,location='bottom')
  plt.show()
```
