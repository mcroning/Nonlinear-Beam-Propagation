```python
# @title Copy path to npz data file from google drive to data_file_name
data_file_name = "/content/drive/MyDrive/test2/data.npz" # @param {type:"string"}
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
from scipy import ndimage
from mpl_toolkits.axes_grid1 import make_axes_locatable
data=np.load(data_file_name,allow_pickle=True)
prdata=data["arr_0"].item()
ixzf=data["arr_2"]
imin_final=data["arr_3"]
imout_final=data["arr_4"]
```

```python
xaper=prdata['xaper']
yaper=prdata['yaper']
xsamp=prdata['xsamp']
ysamp=prdata['ysamp']
gl=prdata['gl']
```

```python
ymax=yaper/2/2 # in mm 1st /2 is for half ya aperture spanning, 2nd /2 is for zero centered yaper
fig,ax=plt.subplots()
ax.set_title(r'input image with gain $\gamma \ell$ '+str(gl))
ax.set_xlabel(r'x ($\mu$m)')
ax.set_ylabel(r'y ($\mu$m)')
img=ax.imshow(imin_final,extent=[-ymax,ymax,-ymax,ymax],vmax=1)
plt.colorbar(img)
plt.show()
```

```python
ymax=yaper/2/2 # in mm 1st /2 is for half ya aperture spanning, 2nd /2 is for zero centered yaper
fig,ax=plt.subplots()
ax.set_title(r'output image with gain $\gamma \ell$ '+str(gl))
ax.set_xlabel(r'x ($\mu$m)')
ax.set_ylabel(r'y ($\mu$m)')
img=ax.imshow(imout_final,extent=[-ymax,ymax,-ymax,ymax],vmax=1)
plt.colorbar(img)
plt.show()
```

```python
fig,ax=plt.subplots()
ixzf=np.where(ixzf<0.0001,0.0001,ixzf)
ax.set_title(r'lognormal scale $\gamma \ell$ '+ str(gl))
ax.set_xlabel(r'x ($\mu$m)')
ax.set_ylabel(r'z ($\mu$m)')
img=ax.imshow(ixzf,norm=colors.LogNorm(vmin=0.01,vmax=ixzf.max()),extent=[-xaper/2,xaper/2,prdata['rlen'],0],aspect=1)
divider = make_axes_locatable(ax)
cax2 = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img,cax=cax2)
plt.show()
```
